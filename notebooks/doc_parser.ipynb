```python
def build_json(data, filename):
    base = filename.split("/")[-1]
    match = re.match(r'(client\d+)_session(\d+)\.txt$', base)
    if match:
        client_id = match.group(1)         # e.g., "client2"
        session_number = int(match.group(2))  # e.g., 2 (as integer)
    else:
        client_id = "NA"
        session_number = "NA"
    condensed_data = {
        "therapy_session_number": session_number,
        "client_id": client_id,
        "chief_complaint": data["Presentation"]['Chief Complaint'],
        "symptoms": [
            {
                "description": symptom['Description'],
                "onset": symptom['Onset'],
                "frequency": symptom['Frequency'],
                "ascendance": symptom['Ascendance'],
                "intensity": symptom['Intensity'],
                "duration": symptom['Duration'],
                "quote": symptom['Quote (Symptom)']
            } for symptom in data["Psychological Factors"]['Symptoms'].values()
        ],

        "diagnosis": [
            {
                "dsm_5_diagnosis": diagnosis['Description'],
                "dsm_5_code": diagnosis.get('DSM- Code', 'NA'),
                "icd_10_code": diagnosis.get('ICD- Code', 'NA'),
            } for diagnosis in data["Clinical Assessment"]['Diagnosiss'].values()
        ],
        "mood_and_affect": data["Mental Status Exam"]['Mood and Affect'],
        "speech_and_language": data["Mental Status Exam"]['Speech and Language'],
        "thought_process_and_content": data["Mental Status Exam"]['Thought Process and Content'],
        "orientation": data["Mental Status Exam"]['Orientation'],
        "perceptual_disturbances": data["Mental Status Exam"]['Perceptual Disturbances'],
        "cognition": data["Mental Status Exam"]['Cognition'],
        "insight": data["Mental Status Exam"]['Insight'],
        "risks_or_safety_concerns": data["Risk Assessment"]['Risks or Safety Concerns'],
        "hopelessness": data["Risk Assessment"]['Hopelessness'],
        "suicidal_thoughts_or_attempts": data["Risk Assessment"]['Suicidal Thoughts or Attempts'],
        "self_harm": data["Risk Assessment"]['Self Harm'],
        "dangerous_to_others": data["Risk Assessment"]['Dangerous to Others'],
        "risk_quote": data["Risk Assessment"]['Quote (Risk)'],
        "response_to_treatment": 'NA' if session_number == 1 else data["Progress and Response"]['Response to Treatment'],
    }
    # convert to string
    condensed_data_str = json.dumps(condensed_data, indent=4)
    return condensed_data, condensed_data_str
```

```python
json_data, json_string = build_json(data, file_name)
```
